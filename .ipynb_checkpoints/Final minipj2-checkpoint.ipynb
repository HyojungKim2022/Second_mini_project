```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#파일 불러옴
education = pd.read_csv('./dataset/education.csv')
```

```python
education.info()
```

### 정보
- Year:2013-2021의 정보를 포함
- State:총 50개주 + 워싱턴 디씨 의 자료 포함
- Type:사립, 주립(주외,주내)
- Length:4년제,2년제
- Expense: 학비, 기숙사비
- Value:가격

# 시각화

```python
df_1 = education.groupby(by=['State']).mean().drop('Year',axis=1)
plot_order = df_1.sort_values(by='Value',ascending=False).index.values
```

```python
# 주별 가격 바플롯
sns.set(font_scale = 2)
plt.figure(figsize=(20,100))
sns.barplot(data=education, x='Value', y='State',order=plot_order)
```

```python
# 주별 기숙사 비용 바플롯
edu = education['Expense'] == 'Room/Board'

a=education[edu]
sns.set(font_scale = 2)
plt.figure(figsize=(20,100))
sns.barplot( x=a.Value,y=a.State)
```

```python
#baxplot
education.boxplot(column=['Value'])
```

```python
# 연도별 총 학비 추이
plt.scatter(education.Year,education.Value)
plt.show()
```

```python
#공립, 사립, 주립별 가격 추이  /사립이 비싸고 주립이저렴/주내와 주외의 학생 가격도 다름
sns.displot(data=education,kde=True,x='Value',hue='Type',palette='Set1',)
plt.show()
```
